```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
import nagisa
import re
```

```python
# Read the CSV file
df=pd.read_csv('preprocess1.csv')
#df.head()
```

```python
# Round1: Remove the emoji
df['Round1']=df['Tweet_no_http'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
```

```python
# Round2: Remove spaces
df['Round2'] = df['Round1'].replace(' ', '', regex=True)
```

```python
# Round3: Remove newlines:\n
df['Round3']=df['Round2'].replace('\n','', regex=True)
```

```python
# Takes in a document, filtering out particles, punctuation, and verb endings

def tokenize_jp(text):
    doc = nagisa.filter(text, filter_postags=['助詞', '補助記号', '助動詞'])
    return doc.words

vectorizer = TfidfVectorizer(
                                tokenizer=tokenize_jp, 
                                max_features=100,
                                max_df=0.8,
                                min_df=5,
                                ngram_range = (1,3),
                                stop_words=stop_words
                                
                            )
matrix = vectorizer.fit_transform(df['Round3'])




words_df = pd.DataFrame(matrix.toarray(),
                        columns=vectorizer.get_feature_names_out())
words_df
```

```python
feature_names = vectorizer.get_feature_names()

dense = matrix.todense()
denselist = dense.tolist()
```

```python
all_keywords = []
for description in denselist:
    x=0
    keywords = []
    for word in description:
        if word > 0:
            keywords.append(feature_names[x])
        x=x+1
    all_keywords.append(keywords)
print (texts[0])
print (all_keywords[0])
```

```python
from sklearn.cluster import KMeans
```

```python
true_k = 5
model = KMeans(n_clusters=true_k, init="k-means++", max_iter=100, n_init=1)
model.fit(matrix)
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

with open ("/Users/danmengcai/Desktop/Folders/PyPractice/japanese_tweets_analysis/results_japanese.txt", "w", encoding="utf-8") as f:
    for i in range(true_k):
        f.write(f"Cluster {i}")
        f.write("\n")
        for ind in order_centroids[i, :10]:
            f.write (' %s' % terms[ind],)
            f.write("\n")
        f.write("\n")
        f.write("\n")
```
